```python
%matplotlib inline
import matplotlib.pyplot as plt
import pyprocar
fig = plt.figure(figsize=(13, 13))

plt.rcParams["mathtext.default"] = "regular"  # Roman ['rm', 'cal', 'it', 'tt', 'sf',
#                                                   'bf', 'default', 'bb', 'frak',
#                                                   'circled', 'scr', 'regular']
plt.rcParams["font.family"] = "Georgia"
plt.rc("font", size=22)  # controls default text sizes
plt.rc("axes", titlesize=22)  # fontsize of the axes title
plt.rc("axes", labelsize=22)  # fontsize of the x and y labels
plt.rc("xtick", labelsize=22)  # fontsize of the tick labels
plt.rc("ytick", labelsize=40)  # fontsize of the tick labels

plt,ax1=pyprocar.bandsplot('PROCAR',outcar='OUTCAR',grid=False,elimit=[-5,5], vmin=-1,vmax=1,knames=['K','$\Gamma$'],cmap='jet',mode='parametric',spin=3)

ax1.set_ylabel(r"Energy [eV]",fontsize=60)
ax1.set_ylim(-0.5,0.5)
ax1.set_xticks([])
ax1.patch.set_edgecolor('black')
ax1.patch.set_linewidth('4') 
plt.tight_layout()
plt.savefig("zoom.png", format="png",dpi=600)
```
